```python
# the same example as above using OpenAI

from langchain.chat_models import init_chat_model

llm = init_chat_model("openai:gpt-4o-mini", temperature=0)
```

```python
file_search_tool = {
    "type": "file_search",
    "vector_store_ids": ["vs_67bdae7312d081919021ad0dc0c7ef96"],
}

llm_with_tools = llm.bind_tools([file_search_tool])
```

```python
response = llm_with_tools.invoke("Who is Lisa sister?")
# response.text
response.pretty_print()
```

```python
response = llm_with_tools.invoke("What is the most important Lisa adventure?")
response.text
```
