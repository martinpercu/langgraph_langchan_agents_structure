```python
import os

from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI

gpt4_mini_llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
response = gpt4_mini_llm.invoke("Hey how is going ?")
response.text
```

```python
# With pretty_print() ===> Will add ===================== Ai Message ====================
# This is very usefull when need to see the messages user AI
response.pretty_print()
```

```python
gpt3_llm = ChatOpenAI(model="gpt-3.5-turbo-0125", temperature=0)
response = gpt3_llm.invoke("Hey how is going ?")
response.pretty_print()
```

```python
# this is the basic setting for agents. What is the llm role etc etc. Giving minimum context.
# The history we send to LLM 
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage

msg_1 = SystemMessage(content="You are a helpful assistant that talks in french")
msg_2 = HumanMessage(content="Je suis Martin D'arc")
msg_3 = AIMessage(content="Bonjour Martin, Comment Ã§a tu?")
msg_4 = HumanMessage(content="Comment je m'appelle?")
history = [msg_1, msg_2, msg_3, msg_4]
response = gpt4_mini_llm.invoke(history)
response.pretty_print()
```

```python
# Using now the Anthropic model 

from langchain_anthropic import ChatAnthropic

claude_llm = ChatAnthropic(model="claude-haiku-4-5-20251001", temperature=0)
response = claude_llm.invoke(history)
response.pretty_print()
```

```python
# With "init_chat_model" langchain will stablish the LLM not matter if is openAI Anthropic or whatever

from langchain.chat_models import init_chat_model

llm = init_chat_model("anthropic:claude-haiku-4-5-20251001", temperature=0)
response = llm.invoke(history)
response.pretty_print()
```

```python
# the same example as above using OpenAI

from langchain.chat_models import init_chat_model

llm = init_chat_model("openai:gpt-4o-mini", temperature=0)
response = llm.invoke(history)
response.pretty_print()
```
