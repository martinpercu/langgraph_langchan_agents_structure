```python
# the AI message
from langchain_core.messages import AIMessage

ai_msg = AIMessage(content="Message from the AI Important remember is an AI message the system needs that")
ai_msg.text
```

```python
# the HUMAN message

from langchain_core.messages import HumanMessage

human_msg = HumanMessage(content="Message from the Human from HUMAN important VERY importnat to know")
human_msg.text
```

```python
# here we keep all the message in the interactions HUMAN==>AI==>HUMAN==>  etc etc etc



history = [ai_msg, human_msg]
for m in history:
    m.pretty_print()
```

```python
# The history is already a lis ==> so to add a new message 

history = history + [AIMessage(content="Message from the AI again AGAIN")]
```

```python
## this will add the last AI message
for m in history:
    m.pretty_print()
```
